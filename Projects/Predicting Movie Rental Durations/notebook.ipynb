```python
df['rental_length_days']=df['rental_length_days'].astype(str).str.replace("days 00:00:00","")
df['rental_length_days']=df['rental_length_days'].astype('int')
```

```python
cols_to_drop=['rental_date','return_date','special_features','rental_length_days']
X=df.drop(cols_to_drop,axis=1)
y=df['rental_length_days']
```

```python
from sklearn.model_selection import train_test_split
```

```python
random_state=9
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=random_state)
```

```python
print(len(X_train))
print(len(y_train))
print(len(X_test))
print(len(y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor 
from sklearn.metrics import mean_squared_error
```

```python
dt=DecisionTreeRegressor()
dt.fit(X_train,y_train)
y_pred_dt=dt.predict(X_test)
mse_dt=mean_squared_error(y_pred_dt,y_test)
print(f"The mean squared error for decision tree is {mse_dt}")
```

```python
rf=RandomForestRegressor()
rf.fit(X_train,y_train)
y_pred_rf=rf.predict(X_test)
mse_rf=mean_squared_error(y_pred_rf,y_test)
print(f"The mean squared error for decision tree is {mse_rf}")
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid = {'n_estimators': np.arange(1,50,1),'max_depth': np.arange(1,20,1)}
rf=RandomForestRegressor()
random_forest_cv=RandomizedSearchCV(estimator=rf,cv=5,param_distributions=param_grid,random_state=random_state)
random_forest_cv.fit(X_train,y_train)
best_params=random_forest_cv.best_params_
print(f"Best Parameters for Tuned Random Forest regressor are {best_params}")
```

```python
rf_final=RandomForestRegressor(**best_params,random_state=random_state)
rf_final.fit(X_train,y_train)
y_pred_final=rf_final.predict(X_test)
mean_squared_error_final=mean_squared_error(y_pred_final,y_test)
print(f"The final mean squared error for Tuned Random Forest is {mean_squared_error_final}")
```
