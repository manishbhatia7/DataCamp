```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('netflix_data.csv')
```

```python
df.head()
```

```python
netflix_subset=df[df['type']=='Movie']
subset=netflix_subset[netflix_subset['release_year']>=1990]
```

```python
movie_1990s=subset[subset['release_year']<2000]
```

```python
plt.hist(movie_1990s['duration'])
plt.title('Distribution of movie durations in 1990s')
plt.xlabel('Duration(In Minutes)')
plt.ylabel('No of movies')
plt.show()
```

```python
duration=100
action_movie_1990s=movie_1990s[movie_1990s['genre']=='Action']
```
